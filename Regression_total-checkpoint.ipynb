```python
# แบ่งข้อมูลออกเป็น train set และ test set
X_train, X_test, y_train, y_test = train_test_split(df[feature_cols], df["%HIline"], test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

# สร้างโมเดล Linear Regression
model = LinearRegression()

# Train โมเดลด้วย X_train และ y_train
model.fit(X_train, y_train)

# แสดงค่า coefficient และ p-value ของแต่ละ feature โดยใช้ statsmodels
X_train = sm.add_constant(X_train) # เพิ่มคอลัมน์ค่าคงที่เข้าไปใน X_train
model_sm = sm.OLS(y_train, X_train).fit()
print(model_sm.summary())
```
