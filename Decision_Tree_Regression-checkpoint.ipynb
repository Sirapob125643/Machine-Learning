```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import (
    confusion_matrix,
    plot_confusion_matrix,
    classification_report,
    accuracy_score,
    precision_score,
    recall_score,
    precision_recall_fscore_support,
    f1_score,
)
```

```python
# à¸à¸³à¸«à¸™à¸” features à¸—à¸µà¹ˆà¸ˆà¸°à¹ƒà¸Šà¹‰ train à¹‚à¸¡à¹€à¸”à¸¥
feature_cols = ["%Hic", "%Hica", "%HIins", "%HIss", "%HIfound", "%Hilp", "%Hitowacc", "%HIrow" ,"%HItower" ]
```

```python
# à¹à¸šà¹ˆà¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸­à¸­à¸à¹€à¸›à¹‡à¸™ train set à¹à¸¥à¸° test set
X_train, X_test, y_train, y_test = train_test_split(df[feature_cols], df["%HIline"], test_size=0.2, random_state=42)
```

```python
# à¸ªà¸£à¹‰à¸²à¸‡à¹‚à¸¡à¹€à¸”à¸¥ Decision Tree Regression à¹à¸¥à¸° train à¸”à¹‰à¸§à¸¢ train set
regr = DecisionTreeRegressor(max_depth=2)
regr.fit(X_train, y_train)
```

```python
# à¸—à¸³à¸™à¸²à¸¢à¸„à¹ˆà¸² %HItower à¸”à¹‰à¸§à¸¢ test set
y_pred = regr.predict(X_test)
```

```python
# à¸„à¸³à¸™à¸§à¸“à¸„à¹ˆà¸² R-squared à¸‚à¸­à¸‡à¹‚à¸¡à¹€à¸”à¸¥
from sklearn.metrics import r2_score

r2_score(y_test, y_pred)
```

```python
new_data = {
    "%Hic": 10,
    "%Hica": 20,
    "%HIins": 30,
    "%HIss": 40,
    "%HIfound": 50,
    "%Hilp": 60,
    "%Hitowacc": 70,
    "%HIrow": 80,
    "%HItower": 80,
}

new_X = pd.DataFrame.from_dict(new_data, orient="index").T

predicted_y = regr.predict(new_X)

print("Predicted %HIline:", predicted_y[0])
```

```python
from sklearn.tree import export_graphviz
import graphviz

# assuming `regr` is already trained
dot_data = export_graphviz(regr, out_file=None, 
                           feature_names=new_X.columns.values, 
                           filled=True, rounded=True, special_characters=True)
graph = graphviz.Source(dot_data)
graph.render("decision_tree")  # save to file
graph.view()  # show in default viewer
```
