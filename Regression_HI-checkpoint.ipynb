```python
# à¹à¸šà¹ˆà¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸­à¸­à¸à¹€à¸›à¹‡à¸™ train set à¹à¸¥à¸° test set
X_train, X_test, y_train, y_test = train_test_split(df[feature_cols], df["%HIline"], test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

# à¸ªà¸£à¹‰à¸²à¸‡à¹‚à¸¡à¹€à¸”à¸¥ Linear Regression
model = LinearRegression()

# Train à¹‚à¸¡à¹€à¸”à¸¥à¸”à¹‰à¸§à¸¢ X_train à¹à¸¥à¸° y_train
model.fit(X_train, y_train)

# à¹à¸ªà¸”à¸‡à¸„à¹ˆà¸² coefficient à¹à¸¥à¸° p-value à¸‚à¸­à¸‡à¹à¸•à¹ˆà¸¥à¸° feature à¹‚à¸”à¸¢à¹ƒà¸Šà¹‰ statsmodels
X_train = sm.add_constant(X_train) # à¹€à¸žà¸´à¹ˆà¸¡à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œà¸„à¹ˆà¸²à¸„à¸‡à¸—à¸µà¹ˆà¹€à¸‚à¹‰à¸²à¹„à¸›à¹ƒà¸™ X_train
model_sm = sm.OLS(y_train, X_train).fit()
print(model_sm.summary())
```
