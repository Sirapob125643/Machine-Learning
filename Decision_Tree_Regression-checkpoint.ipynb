```python
# กำหนด features ที่จะใช้ train โมเดล
feature_cols = ["%Hic", "%Hica", "%HIins", "%HIss", "%HIfound", "%Hilp", "%Hitowacc", "%HIrow" ,"%HItower" ]
```

```python
# แบ่งข้อมูลออกเป็น train set และ test set
X_train, X_test, y_train, y_test = train_test_split(df[feature_cols], df["%HIline"], test_size=0.2, random_state=42)
```

```python
# สร้างโมเดล Decision Tree Regression และ train ด้วย train set
regr = DecisionTreeRegressor(max_depth=2)
regr.fit(X_train, y_train)
```

```python
# ทำนายค่า %HItower ด้วย test set
y_pred = regr.predict(X_test)
```

```python
# คำนวณค่า R-squared ของโมเดล
from sklearn.metrics import r2_score

r2_score(y_test, y_pred)
```

```python
new_data = {
    "%Hic": 10,
    "%Hica": 20,
    "%HIins": 30,
    "%HIss": 40,
    "%HIfound": 50,
    "%Hilp": 60,
    "%Hitowacc": 70,
    "%HIrow": 80,
    "%HItower": 80,
}

new_X = pd.DataFrame.from_dict(new_data, orient="index").T

predicted_y = regr.predict(new_X)

print("Predicted %HIline:", predicted_y[0])
```

```python
from sklearn.tree import export_graphviz
import graphviz

# assuming `regr` is already trained
dot_data = export_graphviz(regr, out_file=None, 
                           feature_names=new_X.columns.values, 
                           filled=True, rounded=True, special_characters=True)
graph = graphviz.Source(dot_data)
graph.render("decision_tree")  # save to file
graph.view()  # show in default viewer
```
